```python
import cv2 as cv
import numpy as np
import sys
import os
import matplotlib.pyplot as plt
```

```python
# Read the image 
img_path = os.path.join('..', 'data', 'birds_lake.jpg')
img = cv.imread(img_path, cv.IMREAD_COLOR)

if img is None:
    sys.exit("Could not read the image.")

cropped_img = img[0:400, 100:500]

plt.imshow(cv.cvtColor(cropped_img, cv.COLOR_BGR2RGB))
```

```python
img_gray = cv.cvtColor(cropped_img, cv.COLOR_BGR2GRAY)

# Apply the thresholding
_, thresh = cv.threshold(img_gray, 45, 255, cv.THRESH_BINARY_INV)

# Find the contours
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

contours_cnt = 0
for cnt in contours:
    if cv.contourArea(cnt) > 10:
        cv.drawContours(cropped_img, [cnt], -1, (0, 255, 0), 1)
        x1, y1, w, h = cv.boundingRect(cnt)
        cv.rectangle(cropped_img, (x1, y1), (x1+w, y1+h), (255, 0, 0), 2)
        contours_cnt += 1
print(f"Number of contours: {contours_cnt}")
# Save the image
cv.imwrite("../submission/birds_lake_contours.jpg", cropped_img)


# Display the image
plt.figure()
plt.imshow(cv.cvtColor(cropped_img, cv.COLOR_BGR2RGB))
plt.show()

# Display the thresholded image
plt.figure()
plt.imshow(thresh, cmap='gray')
plt.show()

# Display the grayscale image
plt.figure()
plt.imshow(img_gray, cmap='gray')
plt.show()
```
